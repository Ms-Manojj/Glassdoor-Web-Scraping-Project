```python
import os
from bs4 import BeautifulSoup

folder_path = r'D:\Main\Python Data Science cource\Web Scraping Projects\Glassdoor\all html'

company_names = []
rating = []
employer_size=[]
industry=[]
review=[]
salaries=[]
jobs=[]
location=[]
description=[]



for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    
    with open(file_path, 'r', encoding='utf-8') as f:
        html = f.read()

# Compnay_name columns

    soup = BeautifulSoup(html, 'html.parser')
    company_name_tags = soup.find_all('h2', {'data-test': 'employer-short-name'})
    
    for company_name_tag in company_name_tags:
        company_names.append(company_name_tag.text)

# Rating_column
    
  
    soup = BeautifulSoup(html, 'html.parser')
    html_content = soup.find_all('b')
    # Extract numerical values from <b> tags
    rating_list = [tag.text for tag in html_content if tag.text.replace('.', '', 1).isdigit()]
    rating.append(rating_list)
    
            
# Employer-size

        

    employer_size_spans = soup.find_all('span', {'class': 'd-block mt-0 css-56kyx5', 'data-test': 'employer-size'})

    # Iterate over the list of matching elements and access the text content
    for employer_size_span in employer_size_spans:
        employer_size_value = employer_size_span.text
        employer_size.append(employer_size_value)

# industry columns

    industry_spans = soup.find_all('span', {'class': 'd-block mt-0 css-56kyx5', 'data-test': 'employer-industry'})

    for industry_span in industry_spans:
        industry_value = industry_span.text
        industry.append(industry_value)        

 
# review_columns


    review_spans = soup.find_all('h3', {'class': 'mt-xsm mt-md-0', 'data-test': 'cell-Reviews-count'})

    for review_span in review_spans:
        review_value = review_span.text
        review.append(review_value)
        
        
# Salaries columns

    Salaries_spans = soup.find_all('h3', {'class': 'mt-xsm mt-md-0', 'data-test': 'cell-Salaries-count'})

    for Salaries_span in Salaries_spans:
        Salaries_value = Salaries_span.text
        salaries.append(Salaries_value)

# Jobs columns

    Jobs_spans = soup.find_all('h3', {'class': 'mt-xsm mt-md-0', 'data-test': 'cell-Jobs-count'})

    for Jobs_span in Jobs_spans:
        Jobs_value = Jobs_span.text
        jobs.append(Jobs_value)


# location columns

    location_spans = soup.find_all('span', {'class': 'd-block mt-0 css-56kyx5', 'data-test': 'employer-location'})

    for location_span in location_spans:
        location_value = location_span.text
        location.append(location_value)  
        

# # description column
  
#     new_spans = soup.find_all('p', {'class': 'css-1sj9xzx css-56kyx5'})
#     for new_span in new_spans:
#         new_value = new_span.text
#         if new_value =='':
#             description.append(new_value)
#         else:
#             description.append(np.nan)
   
         
```

```python
# Using nested loops to access elements one by one
# making new columns for rating
rating_new=[]
for outer_list in rating:
    for element in outer_list:
        rating_new.append(element)
```

```python
len(company_names),len(rating_new),len(employer_size),len(industry),len(review),len(salaries),len(jobs),len(location)
```

```python
# Making a daset for all columns


df = pd.DataFrame({
'company_names':company_names,
'rating_new':rating_new,
'employer_size':employer_size ,
'industry':industry,
'review':review,
'salaries':salaries,
'jobs':jobs,
'location':location 
})
```
